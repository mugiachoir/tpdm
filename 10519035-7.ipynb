```python
#Menggunakan library standar untuk classification
import numpy as np
import pandas as pd
import string
import nltk

#import untuk machine learning naive bayes
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn import model_selection
from sklearn.feature_extraction.text import CountVectorizer,TfidfTransformer
from sklearn.model_selection import train_test_split
from sklearn.utils.multiclass import unique_labels
from sklearn.preprocessing import LabelBinarizer,OrdinalEncoder,OneHotEncoder

#visualisasi hasil
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree


```

```python
#Memanggil dataset yang digunakan
filecsv='bankfull01.csv' 
teks=pd.read_csv(filecsv, header=0, delimiter=';',encoding='utf-8')
df=pd.DataFrame(teks)
print(df)
```

```python
#input parameter untuk prediksi class
xTarget=df.drop(['id','y'],axis=1)
print(xTarget)
```

```python
#target classification dari accepted/rejected loan
yTarget=df['y']
print(yTarget)
```

```python
#Merubah encode nilai yTarget menjadi dua class yaitu 1 atau 0
encoder=LabelBinarizer()
Y=encoder.fit_transform(yTarget)
print(Y)
```

```python
#merubah encode nilai atribut menjadi index nilai
tfidf_transformer=OneHotEncoder()
X= tfidf_transformer.fit_transform(xTarget)
print(X)
print(X.shape)
```

```python
#Membuat data training dan data testing dari dataset dengan data test = 30% dari jumlah instance
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#Melakukan pembuatan model training dengan NaiveBayes
NaiveBayes=MultinomialNB().fit(X_train,np.ravel(y_train,order='C'))
DTree=DecisionTreeClassifier().fit(X_train,np.ravel(y_train,order='C'))
Knn=KNeighborsClassifier(n_neighbors=3).fit(X_train,np.ravel(y_train,order='C'))
RandomForest=RandomForestClassifier(n_estimators=50,max_depth=3).fit(X_train,np.ravel(y_train,order='C'))
SuppVM=SVC(gamma='scale',decision_function_shape='ovo',kernel='linear').fit(X_train,np.ravel(y_train,order='C'))

print(NaiveBayes)
print(DTree)
print(Knn)
print(RandomForest)
print(SuppVM)
```

```python
#Prediksi terhadap model training yang telah dibuat
predictionNB=NaiveBayes.predict(X_test)
accuracies=accuracy_score(y_test,predictionNB)

predictionKNN=Knn.predict(X_test)
accuracies=accuracy_score(y_test,predictionKNN)

predictionRF=RandomForest.predict(X_test)
accuracies=accuracy_score(y_test,predictionRF)

predictionSVM=SuppVM.predict(X_test)
accuracies=accuracy_score(y_test,predictionSVM)

predictionDTree=DTree.predict(X_test)
accuracies=accuracy_score(y_test,predictionDTree)

print(predictionNB)
print(predictionNB)

print(predictionKNN)
print(predictionKNN)

print(predictionRF)
print(predictionRF)

print(predictionSVM)
print(predictionSVM)

print(predictionDTree)
print(predictionDTree)
```

```python
#Menampilkan matrik konfusi
print(confusion_matrix(y_test,predictionNB))
print(confusion_matrix(y_test,predictionKNN))
print(confusion_matrix(y_test,predictionRF))
print(confusion_matrix(y_test,predictionSVM))
```

```python
#Procedure untuk memetakan matrik konfusi
def plot_confusion_matrix(y_true, y_pred, classes,
                         normalize = False,
                         title = None,
                         cmap = plt.cm.Blues):
    if not title:
        if normalize:
            title = 'Normalized confusion matrix'
        else:
            title = 'Confusion matrix, without normalization'

    #Compute confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    #only use the labels that appear in the data
    classes=classes[unique_labels(y_true,y_pred)]
    if normalize:
        cm=cm.astype('float')/cm.sum(axis=1)[:,np.newaxis]
        print("Normalized confusion matrix")
    else:
        print("Confusion matrix, without normalization")
    print(cm)

    fig, ax=plt.subplots()
    im=ax.imshow(cm,interpolation="nearest",cmap=cmap)
    ax.figure.colorbar(im,ax=ax)

    #We want to show all ticks
    ax.set(xticks=np.arange(cm.shape[1]),yticks=np.arange(cm.shape[0]),xticklabels=classes, yticklabels=classes,title=title,ylabel='True label',xlabel='Predicted label')

    #rotate the ticks labels and set their alignment
    plt.setp(ax.get_xticklabels(),rotation=45,ha='right',rotation_mode='anchor')

    #Loop over data dimensions and create text annotation
    fmt='.2f' if normalize else 'd'
    thresh=cm.max()/2
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j,i,format(cm[i,j],fmt),ha='center',va='center',color='white' if cm[i,j] > thresh else 'black')
    fig.tight_layout()

    return ax
```

```python
#Visualisasi Hasil
class_names = yTarget
np.set_printoptions(precision = 2)


print("NAIVE BAYES")
# Plot non-normalized confusion matrix
plot_confusion_matrix(y_test, predictionNB, classes=class_names,
                      title = 'Confusion matrix, without normalization')

# Plot normalized confusion matrix
plot_confusion_matrix(y_test, predictionNB, classes=class_names, normalize=True,
                      title = 'Normalized confusion matrix')

print("KONVOLUSIONAL NEURAL NETWORK")
# Plot non-normalized confusion matrix
plot_confusion_matrix(y_test, predictionKNN, classes=class_names,
                      title = 'Confusion matrix, without normalization')

# Plot normalized confusion matrix
plot_confusion_matrix(y_test, predictionKNN, classes=class_names, normalize=True,
                      title = 'Normalized confusion matrix')

print("RANDOM FOREST")
# Plot non-normalized confusion matrix
plot_confusion_matrix(y_test, predictionRF, classes=class_names,
                      title = 'Confusion matrix, without normalization')

# Plot normalized confusion matrix
plot_confusion_matrix(y_test, predictionRF, classes=class_names, normalize=True,
                      title = 'Normalized confusion matrix')

print("SUPP VM")
# Plot non-normalized confusion matrix
plot_confusion_matrix(y_test, predictionSVM, classes=class_names,
                      title = 'Confusion matrix, without normalization')

# Plot normalized confusion matrix
plot_confusion_matrix(y_test, predictionSVM, classes=class_names, normalize=True,
                      title = 'Normalized confusion matrix')

plt.show()

print("DECISION TREE")
plot_tree(DTree, filled=True)
```
